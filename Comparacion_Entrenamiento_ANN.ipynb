### Comparacion entrenamiento Red con VBLS categoricas y entrenamineto Red sin VBLS categoricas

```python
import matplotlib.pyplot as plt
import numpy as np
import pandas as pd
ANN1=pd.read_csv("h1.csv")

ANN2=pd.read_csv("h2.csv")
plt.figure(figsize=(10, 10))
plt.xlabel("# Epoca")
plt.ylabel("Magnitud de p√©rdida")
plt.plot(ANN1[["0"]], label="ANN sin variables categoricas")
plt.plot(ANN2[["0"]],label="ANN con variables categoricas")
plt.ylim([0, 100])
plt.xlim([0, 140])
plt.legend(loc="upper right")
plt.title("Comparacion de entrenamientos")
plt.show()
```
